```python
import pandas as pd
```

```python
morsels_sales = pd.read_csv("./data/daily_sales_data_2.csv")
```

```python
morsels_sales
```

```python
morsels_pink_price = morsels_sales["price"].str.replace('[\$\,]', '',regex=True).astype(float)
```

```python
morsels_pink_quantity = morsels_sales["quantity"]
```

```python
morsels_sales['sales'] = morsels_pink_price* morsels_pink_quantity
```

```python
morsels_sales
```

```python
morsels_sales.drop('price', axis=1,inplace=True)
```

```python
morsels_sales.drop('quantity', axis=1,inplace=True)
```

```python
morsels_sales
```

```python
morsels_sales[morsels_sales["product"] == "pink morsel"].to_csv("third-modified-data.csv",index=False)
```
